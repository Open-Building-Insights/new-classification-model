### Calculate AREA in meters & PERIMETER in meters

```python
import pandas as pd
import geopandas as gpd
import shapely
from pyproj import Geod

geod = Geod(ellps="WGS84")
```

DATA LOADING

```python
#data obtained from OSM, saved into parquet
df = pd.read_parquet(r"residential.parquet")
gdf = gpd.GeoDataFrame(df, geometry=df["geometry"].apply(shapely.wkb.loads))
print(gdf.columns)
gdf["geometry"]
```

AREA & PERIMETER CALCULATION


```python

def calculate_area_perimeter(geom):
    if geom.is_empty:
        return 0.0, 0.0

    area, perimeter = geod.geometry_area_perimeter(geom)
    return abs(area), perimeter

gdf["area_in_meters"], gdf["perimeter_in_meters"] = zip(*gdf.geometry.apply(calculate_area_perimeter))
gdf["area_in_meters"]
#gdf["perimeter_in_meters"] 
```
